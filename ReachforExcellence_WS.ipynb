# Part 3 - Deep Learning on Cancer Dataset

To do some deep learning we need a lot of premade code because it is way beyond the scope of this session and even the professionals don't write everything out themselves.

```python
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from keras.losses import MeanSquaredError
from sklearn.metrics import confusion_matrix
import tensorflow as tf
import numpy as np
```

Let's define our input features (x) and out output class (Y) from our dataframe. There is a built in bit of code that let's us select the features and then select the target column, seperately. We also then use the np library to convert the dataframe into a big list of features for each cell.

```python
x = np.asarray(df[data.feature_names])
Y_tar =  np.asarray(df['target'])

Y = []
for val in Y_tar:
  if val == 0:
    Y.append([1,0])
  else:
    Y.append([0,1])

Y = np.asarray(Y)
```

Below shows us the dimensions of our data. For our inputs we have 30 features and 569 rows. For our targets we just have one column for each 569 rows.

```python
print(x.shape)
print(Y.shape)
```

Now we need to split the data into a train and test set so that we can see how well the data performs. We'll train on 80% of the data and test on 20% of the data. We'll do this using a function from a package called `sklearn` and specifying the test size as 0.2 (20%).

```python
x_train, x_test, Y_train, Y_test = train_test_split(x, Y, test_size = 0.20, random_state = 0)

print('Number of cells for training = ', x_train.shape[0])
print('Number of cells for testing = ', x_test.shape[0])
```

Now we need to create a model. We'll first create a model that takes all of our inputs, connects this to 1024 hidden nodes, which are connected to another layer of 1024 hidden nodes which are then connected to one output layer.

```python
model = Sequential([
                    Dense(units = 1024, activation = 'relu'),       ## Once you've tried running the code once try changing 1024 to a different number like 16 or 2048
                    Dense(units = 1024, activation = 'relu'),       ## You culd even add more layers by copy and pasting this line as many times as you want.
                    Dense(units = 2, activation = 'sigmoid')
                    ])
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)           ## Try changing the learning rate. Try a lower learning rate like 0.00001 or a higher one like 0.1
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(x_train, Y_train, epochs=50, batch_size=64, verbose = 2, validation_data=(x_test, Y_test))      ##Â Try changing the number of epochs.
```

Now we need to recover the loss so that we can see how the model trained over every epoch.

```python
train_acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
train_loss = history.history['loss']
val_loss = history.history['val_loss']
```

```python
plt.figure(figsize = (16,8))
plt.subplot(1,2,1)
plt.plot(train_acc, linewidth = 3, color = 'tab:blue', label = 'Train')
plt.plot(val_acc, linewidth = 3, color = 'tab:orange', label = 'Val')
plt.legend()
plt.title('Loss Over Training')
plt.xlabel('Epoch')
plt.ylabel('Loss')

plt.subplot(1,2,2)
plt.plot(train_loss, linewidth = 3, color = 'tab:blue', label = 'Train')
plt.plot(val_loss, linewidth = 3, color = 'tab:orange', label = 'Val')
plt.legend()
plt.title('Accuracy Over Training')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.show()
```

We can see that over training the accuracy of the model increased and the loss diecreased. 

Lastly let's see if it how well it performed on the individual classes.

```python
Y_pred = model.predict(x_test)
Y_pred_bin = []
Y_tar_bin = []
for i in range(len(Y_pred)):
  Y_pred_bin.append(np.argmax(Y_pred[i]))
  Y_tar_bin.append(np.argmax(Y_tar[i]))
  
Y_pred_bin = np.asarray(Y_pred_bin)
Y_tar_bin = np.asarray(Y_tar_bin)
```

Now we want to see how the predictions compare to the real answers. We have represented our output in a confusion matrix. These are better explained in a picture [here](https://towardsdatascience.com/confusion-matrix-for-your-multi-class-machine-learning-model-ff9aa3bf7826)

```python
cm = confusion_matrix(Y_tar_bin, Y_pred_bin)
print(cm)
```
